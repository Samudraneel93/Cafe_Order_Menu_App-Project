```python
# Menu Of The Restaurant
menu={
    "Cold Drinks":50,
    "Biryani":120,
    "Momo":60,
    "Fried Potatoes":90,
    "Crinkle Fries":130,
    
}
print(menu)
#Welcome Greetings
print("Welcome to Food_Cafe")
print("Cold_Drinks:Rs 50\nBiryani:Rs 120\nMomo:Rs 60\nFried_Potatoes:Rs 90\nCrinkle_Fries:Rs 130")
order_total=0

item1=input("Enter the name of the item you want to order: " )
if item1 in menu:
    order_total+=menu[item1]
    print("Your Item",item1,"has been added to the order")
else:
    print("The ordered Item",item1,"is not available")
    
another_order=input("Do you want to add another item? Yes/No")
if another_order=="Yes":
    item2= input("Enter the name of the second item:")
    if item2 in menu:
        order_total+=menu[item2]
        print("Your item",item2,"has been added to order")
    else:
        print("The Ordered Item",item2,"is not available")
print("The Total ordered amount to pay:",order_total)
    
```
